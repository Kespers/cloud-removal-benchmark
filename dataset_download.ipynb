```python
import os
import tarfile
import urllib.request

# Definizione del path (assicurati di aver definito dataset_path prima)
dataset_path = "dataset" 

# 3. Crea la struttura delle cartelle se non esiste
if not os.path.exists(dataset_path):
    os.makedirs(dataset_path)
    print(f"Cartella {dataset_path} creata.")

# 4. Spostati nella cartella dataset
os.chdir(dataset_path)

def download_and_extract(url, filename):
    if not os.path.exists(filename):
        print(f"Scaricamento di {filename}...")
        urllib.request.urlretrieve(url, filename)
    
    print(f"Estrazione di {filename}...")
    with tarfile.open(filename, "r:gz") as tar:
        tar.extractall()
    
    # Opzionale: rimuovi il file tar.gz per risparmiare spazio
    # os.remove(filename)

# 5. Scarica ed estrai S1
url_s1 = "ftp://m1659251:m1659251@dataserv.ub.tum.de/s1_asiaWest_test.tar.gz"
download_and_extract(url_s1, "s1_asiaWest_test.tar.gz")

# 6. Scarica ed estrai S2
url_s2 = "ftp://m1659251:m1659251@dataserv.ub.tum.de/s2_asiaWest_test.tar.gz"
download_and_extract(url_s2, "s2_asiaWest_test.tar.gz")

print("\nStruttura creata. Contenuto della cartella:")
print(os.listdir('.'))
```
